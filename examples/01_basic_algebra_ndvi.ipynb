# 01. Basic Band Algebra (NDVI)
This notebook demonstrates how to load a local GeoTIFF, perform a simple band math calculation like the Normalized Difference Vegetation Index (NDVI), and save the result to a new file.

```python
import curaster
import os

# Set paths
data_dir = "../build/benchmark_data"
input_file = os.path.join(data_dir, "test_S_2048x2048.tif")
output_file = "ndvi_output.tif"

print(f"Input file: {input_file}")
```

## Perform Algebra
Assuming Band 5 is Near-Infrared (NIR) and Band 4 is Red. Note: `curaster` uses 1-indexed bands (B1, B2, etc.).
The NDVI formula is `(NIR - Red) / (NIR + Red)`.

```python
# Construct the lazy chain
chain = curaster.open(input_file) \
    .algebra("(B5 - B4) / (B5 + B4)")

# Get info before executing
info = chain.get_info()
print("Output Raster Info:", info)
```

## Save the Result
Calling `.save_local()` will execute the pipeline on the GPU and save a float32 tiled GeoTIFF.

```python
%%time
# Execute and save
chain.save_local(output_file, verbose=True)
print(f"Successfully saved to {output_file}")
```

You can now load this file using any standard GIS software, or verify it using `rasterio`.
